```python
students = students.assign(passed_math = students.math_score >= 60)
students = students.assign(passed_reading = students.reading_score >= 60)
```

```python
students.head()
```

```python
schools_avgs = students[['school_name', 'reading_score', 'math_score']].groupby('school_name').mean()
schools_avgs
```

```python
schools_merged = pd.merge(schools, schools_avgs, on = 'school_name')
schools_merged.head()
```

```python
schools_passing_students = students[['school_name', 'passed_math', 'passed_reading']].groupby('school_name').sum()
schools_passing_students.head()
```

```python
schools_merged = pd.merge(schools_merged, schools_passing_students, on = 'school_name')
schools_merged.head()
```

```python
math_passed_perc    = schools_merged[['passed_math', 'size']].apply(lambda row: 100 * row[0] / row[1], axis=1)
reading_passed_perc = schools_merged[['passed_reading', 'size']].apply(lambda row: 100 * row[0] / row[1], axis=1)
```

```python
schools_merged = schools_merged.assign(passed_math_perc = math_passed_perc,
                                       passed_read_perc = reading_passed_perc)
schools_merged.head()
```

```python
# Create a new column (as a list)
#overall_passed_perc = schools_merged[['passed_math_perc', 'passed_read_perc']].apply(lambda row: (row[0] + row[1])/2, axis=1)
overall_passed_perc = schools_merged[['passed_math_perc', 'passed_read_perc']].apply(np.mean, axis=1)

# Add the list to the dataframe as a column
schools_merged = schools_merged.assign(passed_overall_perc = overall_passed_perc)
schools_merged.head()
```

```python
overall_passed_perc
```

```python
# Per student budget
per_student_budget = schools_merged[['budget', 'size']].apply(lambda row: row[0] / row[1], axis=1)
per_student_budget
```

```python
# Add the list to the dataframe as a column
schools_merged = schools_merged.assign(per_student_budget = per_student_budget)
schools_merged.head()
```


### Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)


```python
schools_merged.head()
```

```python
schools_merged.per_student_budget.hist()
```

```python
schools_merged.per_student_budget.describe()
```

```python
# Spending bins

# bin1 578 - 598
# bin2 599 - 618
# bin3 619 - 638
# bin4 639 - 655

def choose_spending_bin(value): # scope of the function
    if 578 <= value <= 598: # scope of the if statement
        return 'bin1'
    elif 599 <= value <= 618:
        return 'bin2'
    elif 619 <= value <= 638:
        return 'bin3'
    elif 639 <= value:
        return 'bin4'
    
```

```python
schools_merged.per_student_budget.plot('box')
```

```python
schools_merged.per_student_budget.apply(choose_spending_bin)
```

```python
new_columns = ['reading_score', 'math_score',
               'passed_math_perc', 'passed_read_perc',
               'passed_overall_perc', 'per_student_budget']

schools_by_spending = schools_merged[new_columns]

# Creating list of bins
spending_bins = schools_by_spending.per_student_budget.apply(choose_spending_bin)

# Adding list of bins to new dataframe
schools_by_spending = schools_by_spending.assign(spending_bin = spending_bins).drop('per_student_budget', axis=1)
schools_by_spending
```

```python
schools_by_spending.groupby('spending_bin').mean()
```

### Scores by School Size

* Repeat the above breakdown, but this time group schools based on a reasonable approximation of school size (Small, Medium, Large).


```python
schools_merged['size'].describe()
```

```python
schools_merged['size'].hist()
```

```python
5000./3
```

```python
# Size bins

# bin1 0 - 1500
# bin2 1501 - 3000
# bin3 3001 - 5000

def choose_size_bin(value): # scope of the function
    if value <= 1500: # scope of the if statement
        return 'small'
    elif 1501 <= value <= 3000:
        return 'medium'
    elif 3001 <= value:
        return 'large'
    
    
```

```python
new_columns = ['reading_score', 'math_score',
               'passed_math_perc', 'passed_read_perc',
               'passed_overall_perc', 'size']

schools_by_size = schools_merged[new_columns]

# Creating list of bins
size_bins = schools_by_size['size'].apply(choose_size_bin)

# Adding list of bins to new dataframe
schools_by_size = schools_by_size.assign(size_bin = size_bins).drop('size', axis=1)
schools_by_size
```

```python
schools_by_size.groupby('size_bin').mean()
```
